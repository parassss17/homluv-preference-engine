# 01 - Explore the data

HomLuv is a visual home-discovery platform: buyers like/dislike room photos and the AI matches them to a builder by style, price and location. This notebook just looks at the public sample dataset (price, location, sample photos) before we build the recommender in notebook 02.

Run `python prepare_data.py` first so `artifacts/listings.csv` exists.

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image

listings = pd.read_csv('../artifacts/listings.csv')
print(listings.shape)
listings.head()
```

```python
listings.describe()[['bedrooms', 'bathrooms', 'area', 'price']]
```

### Price distribution
Most homes cluster at the lower end with a long tail of expensive ones - worth knowing for the budget filter in the app.

```python
plt.figure(figsize=(7, 3))
plt.hist(listings['price'], bins=40)
plt.xlabel('Price ($)'); plt.ylabel('Homes'); plt.title('Price distribution')
plt.show()
```

### Simulated builders
The public sample has no builder column, so `prepare_data.py` grouped homes by style + price into builder portfolios. Here is how many homes each got.

```python
listings['builder'].value_counts().sort_index()
```

### A few sample homes

```python
fig, axes = plt.subplots(1, 4, figsize=(14, 3))
for ax, (_, row) in zip(axes, listings.head(4).iterrows()):
    img = Image.open(os.path.join('../artifacts/images_sample', row['image']))
    ax.imshow(img); ax.axis('off')
    ax.set_title(f"${int(row['price']):,}\n{row['builder']}")
plt.show()
```
